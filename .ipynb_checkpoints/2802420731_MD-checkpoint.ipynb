# Feature Engineering

## Drop Kolom yang Tidak Berguna

Karena kolom student ID memiliki kardinalitas yang tinggi (banyak nilai unik), maka akan di-drop.

```python
df = df.drop(columns = ["Student_ID"])

df.info()
```

## Encode Categorical Variables

Kolom gender dan branch bersifat nominal, maka akan di-one hot encoding.

```python
cat_cols = [
    "gender",
    "branch",
]

for col in cat_cols:
    df = pd.get_dummies(df, columns = [col], prefix = col, dtype = float)

df.info()
```

Kolom kategorik lain bersifat ordinal, sehingga value masing-masing kolom di-encode dengan bilangan.

```python
df["part_time_job"].value_counts()
```

```python
df["part_time_job"] = df["part_time_job"].replace("No", 0)
df["part_time_job"] = df["part_time_job"].replace("Yes", 1)

df["part_time_job"].value_counts()
```

```python
df["family_income_level"].value_counts()
```

```python
df["family_income_level"] = df["family_income_level"].replace("Low", 0)
df["family_income_level"] = df["family_income_level"].replace("Medium", 1)
df["family_income_level"] = df["family_income_level"].replace("High", 2)

df["family_income_level"].value_counts()
```

```python
df["city_tier"].value_counts()
```

```python
df["city_tier"] = df["city_tier"].replace("Tier 1", 0)
df["city_tier"] = df["city_tier"].replace("Tier 2", 1)
df["city_tier"] = df["city_tier"].replace("Tier 3", 2)

df["city_tier"].value_counts()
```

```python
df["internet_access"].value_counts()
```

```python
df["internet_access"] = df["internet_access"].replace("No", 0)
df["internet_access"] = df["internet_access"].replace("Yes", 1)

df["internet_access"].value_counts()
```

```python
df["extracurricular_involvement"].value_counts()
```

```python
df["extracurricular_involvement"] = df["extracurricular_involvement"].replace("Low", 0)
df["extracurricular_involvement"] = df["extracurricular_involvement"].replace("Medium", 1)
df["extracurricular_involvement"] = df["extracurricular_involvement"].replace("High", 2)

df["extracurricular_involvement"].value_counts()
```

```python
df["placement_status"].value_counts()
```

```python
df["placement_status"] = df["placement_status"].replace("Not Placed", 0)
df["placement_status"] = df["placement_status"].replace("Placed", 1)

df["placement_status"].value_counts()
```

```python
df.info()
```

## Feature Engineering

Akan ditambahkan kolom kemampuan total dan nilai akademik agar model melihat fitur yang bersifat general, sehingga mengurangi noise, membuat pola yang lebih jelas, dan kemungkinan memperbaiki kinerja model. 

```python
df["total_skills"] = (df["coding_skill_rating"] + df["communication_skill_rating"] + 
                      df["aptitude_skill_rating"])
df["academic_score"] = ((df["tenth_percentage"] + df["twelfth_percentage"] + (df["cgpa"] * 10))
                       / 3)

df.info()
```

# Split Features and Target

Karena tugasnya ada 2, maka ada data output yang terpisah untuk klasifikasi (placement status) dan regresi (gaji).

```python
x = df.drop(columns = ["placement_status", "salary_lpa"])
y_class = df["placement_status"]
y_reg = df["salary_lpa"]

x.info()
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train_class, x_test_class, y_train_class, y_test_class = train_test_split(x, y_class, 
                                                                            test_size = 0.2, 
                                                                            random_state = 42)
x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(x, y_reg, test_size = 0.2,
                                                                    random_state = 42)

x_train_class.info()
```

```python
x_test_class.info()
```

```python
x_train_reg.info()
```

```python
x_test_reg.info()
```

# Scaling

Fitur di-scale agar pelatihan model lebih seimbang, adil, cepat, dan akurat.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_class = scaler.fit_transform(x_train_class)
x_test_class = scaler.fit_transform(x_test_class)
x_train_reg = scaler.fit_transform(x_train_reg)
x_test_reg = scaler.fit_transform(x_test_reg)
```

# Pembangunan Algoritma Klasifikasi

## XGBoost

```python
import xgboost as xgb
from sklearn.metrics import classification_report

xgb_class_base = xgb.XGBClassifier()

xgb_class_base.fit(x_train_class, y_train_class)

y_pred_xgb_class_base = xgb_class_base.predict(x_test_class)

print("Base Classification Report - XGBoost Classification")
print(classification_report(y_test_class, y_pred_xgb_class_base, 
                            target_names = ["Not Placed", "Placed"]))
```

```python
from sklearn.model_selection import GridSearchCV

xgb_class_param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.05, 0.1],
    "max_depth": [3, 5],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    "min_child_weight": [1, 3]
}
xgb_class_grid = GridSearchCV(xgb_class_base, xgb_class_param_grid, cv = 5, 
                              scoring = "accuracy", n_jobs = -1)

xgb_class_grid.fit(x_train_class, y_train_class)
```

```python
print("Best Parameters:", xgb_class_grid.best_params_)
print("Best Score:", xgb_class_grid.best_score_)
```

```python
xgb_class_best = xgb.XGBClassifier(
    colsample_bytree = 1.0,
    learning_rate = 0.05,
    max_depth = 3,
    min_child_weight = 1,
    n_estimators = 100,
    subsample = 1.0
)

xgb_class_best.fit(x_train_class, y_train_class)

y_pred_xgb_class_best = xgb_class_best.predict(x_test_class)

print("Best Classification Report - XGBoost Classification")
print(classification_report(y_test_class, y_pred_xgb_class_best, 
                            target_names = ["Not Placed", "Placed"]))
```

## LightGBM

```python
import lightgbm as lgb

lgb_class_base = lgb.LGBMClassifier()

lgb_class_base.fit(x_train_class, y_train_class)

y_pred_lgb_class_base = lgb_class_base.predict(x_test_class)

print("Base Classification Report - LightGBM Classification")
print(classification_report(y_test_class, y_pred_lgb_class_base, 
                            target_names = ["Not Placed", "Placed"]))
```

```python
lgb_class_param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.05, 0.1],
    "num_leaves": [31, 50],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}
lgb_class_grid = GridSearchCV(lgb_class_base, lgb_class_param_grid, cv = 5, 
                              scoring = "accuracy", n_jobs = -1)

lgb_class_grid.fit(x_train_class, y_train_class)
```

```python
print("Best Parameters:", lgb_class_grid.best_params_)
print("Best Score:", lgb_class_grid.best_score_)
```

```python
lgb_class_best = lgb.LGBMClassifier(
    colsample_bytree = 0.8,
    learning_rate = 0.1,
    n_estimators = 100,
    num_leaves = 31,
    subsample = 0.8
)

lgb_class_best.fit(x_train_class, y_train_class)

y_pred_lgb_class_best = lgb_class_best.predict(x_test_class)

print("Best Classification Report - LightGBM Classification")
print(classification_report(y_test_class, y_pred_lgb_class_best, 
                            target_names = ["Not Placed", "Placed"]))
```

## CatBoost

```python
pip install catboost
```

```python
from catboost import CatBoostClassifier

cb_class_base = CatBoostClassifier(verbose = 0)

cb_class_base.fit(x_train_class, y_train_class)

y_pred_cb_class_base = cb_class_base.predict(x_test_class)
```

```python
print("Base Classification Report - CatBoost Classification")
print(classification_report(y_test_class, y_pred_cb_class_base, 
                            target_names = ["Not Placed", "Placed"]))
```

```python
from sklearn.model_selection import RandomizedSearchCV

cb_class_param_grid = {
    "iterations": [100, 200, 300],
    "learning_rate": [0.03, 0.05, 0.1],
    "depth": [4, 6, 8],
    "l2_leaf_reg": [3, 5, 7]
}
cb_class_grid = RandomizedSearchCV(cb_class_base, param_distributions = cb_class_param_grid,
                                   n_iter = 10, cv = 3, scoring = "accuracy", n_jobs = -1,
                                  random_state = 42)

cb_class_grid.fit(x_train_class, y_train_class)
```

```python
print("Best Parameters:", cb_class_grid.best_params_)
print("Best Score:", cb_class_grid.best_score_)
```

```python
cb_class_best = CatBoostClassifier(
    learning_rate = 0.05,
    l2_leaf_reg = 3,
    iterations = 100,
    depth = 6,
    verbose = 0
)

cb_class_best.fit(x_train_class, y_train_class)

y_pred_cb_class_best = cb_class_best.predict(x_test_class)

print("Best Classification Report - CatBoost Classification")
print(classification_report(y_test_class, y_pred_cb_class_best, 
                            target_names = ["Not Placed", "Placed"]))
```

```python
from sklearn.metrics import roc_auc_score

y_prob_xgb = xgb_class_best.predict_proba(x_test_class)[:, 1]
xgb_score = roc_auc_score(y_test_class, y_prob_xgb)
y_prob_lgb = lgb_class_base.predict_proba(x_test_class)[:, 1]
lgb_score = roc_auc_score(y_test_class, y_prob_lgb)
y_prob_cb = cb_class_base.predict_proba(x_test_class)[:, 1]
cb_score = roc_auc_score(y_test_class, y_prob_cb)

print(f"XGBoost ROC AUC Score: {xgb_score}")
print(f"LightGBM ROC AUC Score: {lgb_score}")
print(f"CatBoost ROC AUC Score: {cb_score}")
```

```python
from sklearn.metrics import roc_curve

fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test_class, y_prob_xgb)

plt.plot(fpr_xgb, tpr_xgb)
plt.plot([0, 1], [0, 1], linestyle = "--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("XGBoost ROC Curve")
plt.show()
```

```python
fpr_lgb, tpr_lgb, thresholds_lgb = roc_curve(y_test_class, y_prob_lgb)

plt.plot(fpr_lgb, tpr_lgb)
plt.plot([0, 1], [0, 1], linestyle = "--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("LightGBM ROC Curve")
plt.show()
```

```python
fpr_cb, tpr_cb, thresholds_cb = roc_curve(y_test_class, y_prob_cb)

plt.plot(fpr_cb, tpr_cb)
plt.plot([0, 1], [0, 1], linestyle = "--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("CatBoost ROC Curve")
plt.show()
```

Nilai AUC ketiga algoritma ini sama-sama bagus, namun paling bagus nilai CatBoost, sehingga dapat disimpulkan bahwa CatBoost dapat membedakan class negatif dan class positif dengan baik, meskipun tidak sempurna.

Dengan demikian, algoritma CatBoost akan dipilih untuk tugas klasifiaksi placement status ini.

# Regresi

## XGBoost

```python
xgb_reg_base = xgb.XGBRegressor()

xgb_reg_base.fit(x_train_reg, y_train_reg)

y_pred_xgb_reg_base = xgb_reg_base.predict(x_test_reg)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

print("XGBoost Regression Base:")
print(f"MAE: {mean_absolute_error(y_test_reg, y_pred_xgb_reg_base)}")
print(f"MSE: {mean_squared_error(y_test_reg, y_pred_xgb_reg_base)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test_reg, y_pred_xgb_reg_base))}")
print(f"R2 Score: {r2_score(y_test_reg, y_pred_xgb_reg_base)}")
```

```python
xgb_reg_param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.05, 0.1],
    "max_depth": [3, 5],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    "min_child_weight": [1, 3]
}
xgb_reg_grid = GridSearchCV(xgb_reg_base, xgb_reg_param_grid, cv = 5, 
                              scoring = "neg_mean_squared_error", n_jobs = -1)

xgb_reg_grid.fit(x_train_reg, y_train_reg)
```

```python
print("Best Parameters:", xgb_reg_grid.best_params_)
print("Best Score:", xgb_reg_grid.best_score_)
```

```python
xgb_reg_best = xgb.XGBRegressor(
    colsample_bytree = 0.8,
    learning_rate = 0.05,
    max_depth = 3,
    min_child_weight = 3,
    n_estimators = 200,
    subsample = 0.8
)

xgb_reg_best.fit(x_train_reg, y_train_reg)

y_pred_xgb_reg_best = xgb_reg_best.predict(x_test_reg)

print("XGBoost Regression Best:")
print(f"MAE: {mean_absolute_error(y_test_reg, y_pred_xgb_reg_best)}")
print(f"MSE: {mean_squared_error(y_test_reg, y_pred_xgb_reg_best)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test_reg, y_pred_xgb_reg_best))}")
print(f"R2 Score: {r2_score(y_test_reg, y_pred_xgb_reg_best)}")
```

## CatBoost

```python
from catboost import CatBoostRegressor

cb_reg_base = CatBoostRegressor(verbose = 0)

cb_reg_base.fit(x_train_reg, y_train_reg)

y_pred_cb_reg_base = cb_reg_base.predict(x_test_reg)

print("CatBoost Regression Base:")
print(f"MAE: {mean_absolute_error(y_test_reg, y_pred_cb_reg_base)}")
print(f"MSE: {mean_squared_error(y_test_reg, y_pred_cb_reg_base)}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test_reg, y_pred_cb_reg_base))}")
print(f"R2 Score: {r2_score(y_test_reg, y_pred_cb_reg_base)}")
```

```python
cb_reg_param_grid = {
    "iterations": [100, 200, 300],
    "learning_rate": [0.03, 0.05, 0.1],
    "depth": [4, 6, 8],
    "l2_leaf_reg": [3, 5, 7],
    "bagging_temperature": [0, 0.5, 1]
}
cb_reg_grid = RandomizedSearchCV(cb_reg_base, param_distributions = cb_reg_param_grid,
                                 n_iter = 10, cv = 3, scoring = "neg_mean_squared_error", 
                                 n_jobs = -1, random_state = 42)

cb_reg_grid.fit(x_train_reg, y_train_reg)
```
